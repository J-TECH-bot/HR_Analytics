```python
# we can drop the duplicate data because they are already present in a dataset.
df = df.drop_duplicates()
df
```

```python
#Check Outliners using box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['age'], vert=False)
plt.xlabel('Age')
plt.title('Box Plot for Age (Diabetes Dataset)')
plt.show()
```

```python
# Correlation(Relationship between two columns)
correlation = df.corr()
correlation
```

```python
# To show correlation we are using heat map
# for sowing values we are using annot
#for changing color we use cmap
sns.heatmap(correlation,annot = True,cmap = 'Blues')
```

```python
DV = ['dept', 'numberOfProjects', 'timeSpent.company', 'workAccident', 'promotionInLast5years', 'salary']
DV
```

```python
fig = plt.subplots(figsize=(20,25)) # Size of the figure
for i,j in enumerate(DV): # It is a built-in function in Python that allows you to loop over an iterable object and keep track of the index of each item
    print(i,j)
    plt.subplot(3,2,i+1) #(row,column,i=1(position))Devide the page
        
    sns.countplot(x=j, data=df, hue='left') # j is holding column names, df(data), hue(indicators)
        
    plt.xticks(rotation=90) # to rotate the values of x axis at 90 degree angle.
plt.show()    
```
